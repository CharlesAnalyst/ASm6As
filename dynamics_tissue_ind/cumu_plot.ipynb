```python
import random
##
os.chdir("/Charles/project/ASm6A/ASm6A/cal_variance/cum_plot/")
df = pd.read_table("ASm6A_AR_matrix.txt", index_col=0)
##
def generate_sample_dict():
    sample_list, sample_dict = [], {}
    for ind, i_list in ind_dict.items():
        sample_list += i_list
    tissue_list = [tissue for sample in sample_list for tissue in tissue_dict if sample in tissue_dict[tissue]]
    ind_list = [ind for ind in ind_dict for sample in sample_list if sample in ind_dict[ind]]
    assert len(sample_list) == len(tissue_list) == len(ind_list)
    for x in sample_list:
        sample_dict[x] = {"ind":"", "tissue":""}
    for i in range(len(sample_list)):
        sample_dict[sample_list[i]]['ind'] = ind_list[i]
        sample_dict[sample_list[i]]['tissue'] = tissue_list[i]
    return sample_dict

def select_controls(asm6a, ind, samples): # samples: different tissue samples for one specific individual
    
    def select_one(asm6a, tissue, exclude_ind_list):
        samples = tissue_dict[tissue]
        tmp_samples = list(df.loc[:, samples][df.index==asm6a].dropna(axis=1).columns)
        if len(tmp_samples) == 0:
            return ""
        else:
            valid_samples = [sample for sample in tmp_samples if sample_dict[sample]['ind'] not in exclude_ind_list]
            if len(valid_samples) > 0:
                selected_sample = random.sample(valid_samples,1)[0]
                return selected_sample
            else:
                return ""
            
    cont_list, exclude_ind_list = [], [ind]
    for sample in samples:
        tissue = sample_dict[sample]['tissue']
        i_sample = select_one(asm6a, tissue, exclude_ind_list)
        cont_list.append(i_sample)
        if i_sample != "":
            i_ind = sample_dict[i_sample]['ind']
            exclude_ind_list.append(i_ind)
    nonzero_cont_list = [x for x in cont_list if x != ""]
    if len(nonzero_cont_list)<3:
        return ""
    else: ### control samples only need to >= 3;
        assert len(cont_list) == len(samples)
        index_list = [i for i in range(len(cont_list)) if cont_list[i] != ""]
        case_list = [samples[i] for i in index_list]
        result_list = [case_list, nonzero_cont_list]
        return result_list

##
case_var_list, cont_var_list = [], []
sample_dict = generate_sample_dict()
for ind in list(ind_dict.keys()):
    sample_list = ind_dict[ind]
    selected_asm6as = [term for term,ar_list in df[sample_list].iterrows() if (len(ar_list.dropna())>=3)]
    selected_samples = [list(df.loc[asm6a,sample_list].dropna().index) for asm6a in selected_asm6as]
    assert len(selected_asm6as) == len(selected_samples)
    asm6a_dict = dict(zip(selected_asm6as, selected_samples)) #  ASm6A and cooresponding samples
    ##
    for asm6a, case_samples in asm6a_dict.items():
        result_list = select_controls(asm6a, ind, case_samples)
        if result_list != "":
            case_samples, cont_samples = result_list[0], result_list[1]
#             print(asm6a, case_samples, cont_samples)
            case_ar_list,cont_ar_list=[df.loc[asm6a,x] for x in case_samples],[df.loc[asm6a,x] for x in cont_samples]
            case_var, cont_var = np.var(case_ar_list), np.var(cont_ar_list)
            case_var_list.append(case_var)
            cont_var_list.append(cont_var)
print(len(case_var_list), len(cont_var_list))


#### output file
def write2file(in_list, result_file):
    with open(result_file, 'w') as fw:
        fw.writelines(["%s\n" % str(x) for x in in_list])

result_dir = "/Charles/project/ASm6A/ASm6A/cal_variance/cum_plot/compare_tissue/"
os.system("mkdir -p %s" % result_dir)
write2file(case_var_list, os.path.join(result_dir, "case_variances.txt"))
write2file(cont_var_list, os.path.join(result_dir, "cont_variances.txt"))
```

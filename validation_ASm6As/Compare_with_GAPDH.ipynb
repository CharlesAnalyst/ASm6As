```python
#### To determin it the IP was successful (generation of m6A peak)
#### The test criteria are by comparison with the control (6 interval: GAPDH + OCT4)
#### measure the m6A level of each m6A peak which coveraging the ASm6A sites.

os.chdir("/Charles/project/ASm6A/data/validation_ASm6A/LBFC20201232/bam/")
### bam count (total reads)
input_count = 1956911
ip_count = 3182714
###
count = "/Charles/project/ASm6A/data/validation_ASm6A/LBFC20201232/bam/bamCount.txt"
df = pd.read_table(count)
# input_count = df['MIXinput2'].sum()
# ip_count = df['ASIP'].sum()
# print(input_count, ip_count)
df['length'] = df['END'] - df['START'] + 1
df['length'] = df['length'].astype(int)
df['RPKM_MIXinput2'] =  ((df['MIXinput2']+1)*1000000) / (input_count*df['length'])
df['RPKM_ASIP'] = ((df['ASIP']+1)*1000000) / (ip_count*df['length'])
df['m6A_ratio'] = df['RPKM_ASIP'] / df['RPKM_MIXinput2']
### 内参 GAPDH (3) + OCT4 (3)
start_list = [6643999, 6646151, 6646877, 31132114, 31136797, 31137970]
end_list = [6646160, 6646893, 6647336, 31132581, 31137642, 31138458]
df_control = df[(df['START'].isin(start_list)) & (df['END'].isin(end_list))]
print(df_control)
m6A_ratio = df_control['m6A_ratio'].tolist()
control_ratio = np.mean(m6A_ratio)
print("control ratio: %.2f" % control_ratio)
###
df_query = df[~df['START'].isin(start_list)]
df_sig = df_query[df_query['m6A_ratio'] >= control_ratio]
df_sig = df_sig.sort_values(['m6A_ratio'], ascending=False)
print(len(df_query), len(df_sig))
print(df_sig.head(n=10))
df_failure = df_query[df_query['m6A_ratio'] < control_ratio]
print(df_failure)
```
